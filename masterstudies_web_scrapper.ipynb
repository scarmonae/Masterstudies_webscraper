```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime

def deadline_format(deadline):
    try:
        deadline_obj = datetime.strptime(deadline, "%d %b %Y")
        deadline_formateada = deadline_obj.strftime("%d/%m/%Y")
    except:
        print(f'Error al convertir la fecha {deadline}')
        return ""    
    
    return deadline_formateada

def web_scraping(url):
    # Realizar solicitud HTTP a la página
    try:
        response = requests.get(url)
    except:
        print(f'Error al cargar la página {url}')
        return None
    

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        # Encontrar los elementos específicos en la página
        nombre_programa_tag = soup.find('h1', class_='text-md max-w-3xl font-bold md:text-4xl')
        universidad_tag = soup.find('p', class_='block text-xs mt-2 md:text-base')
        location_tag = soup.find('p', class_='text-base')
        varios_elementos = soup.find_all('p', class_='text-base mt-0.5 first-of-type:font-bold')

        if nombre_programa_tag and universidad_tag:
            nombre_programa = nombre_programa_tag.text.strip()
            universidad = universidad_tag.text.strip()
            location = location_tag.text.strip()
            elementos = [elemento.text.strip() for elemento in varios_elementos]
            idioma = elementos[0]
            modalidad = elementos[1]
            matricula = elementos[4]
            deadline = deadline_format(elementos[5])

            # Obtener la URL de la maestría (si está disponible)
            url_programa = response.url

            # Crear un DataFrame con la fila actual
            df = pd.DataFrame({
                'Nombre del programa': [f'=HYPERLINK("{url_programa}", "{nombre_programa}")'],
                'Universidad': [universidad],
                'Ubicación': [location],
                'Idioma': [idioma],
                'Modalidad': [modalidad],
                'Costo de la matricula': [matricula],
                'Fecha de cierre': [deadline]
            })

            return df
        else:
            print('Elementos no encontrados en la página.')
            return None
    else:
        print('Error al cargar la página:', response.status_code)
        return None

def guardar_excel(df, nombre_archivo):
    # Guardar el DataFrame en un archivo Excel

    columnas_ordenadas = ['Rankeo', 'Calificación personal', 'Aplicación', 'Anotaciones', 'Nombre del programa', 'Universidad', 'Ubicación', 'Fecha de cierre', 'Requisitos que no tengo', 'Costo de la aplicación', 'Costo de la matricula', 'TOEFL', 'Idioma', 'Modalidad']
    df = df[columnas_ordenadas]
    
    df.to_excel(f'{nombre_archivo}.xlsx', index=False, engine='xlsxwriter')
    print(f'Datos guardados en {nombre_archivo}.xlsx')

if __name__ == "__main__":
    # Pedir al usuario que ingrese las URLs hasta que ingrese una cadena vacía
    urls = []
    while True:
        url = input('Ingresa la URL de la maestría (o presiona Enter para terminar): ')
        if not url:
            break
        urls.append(url)

    # Crear un DataFrame vacío con las columnas deseadas
    columnas = ['Rankeo', 'Calificación personal', 'Aplicación', 'Anotaciones', 'Nombre del programa', 'Universidad', 'Ubicación', 'Fecha de cierre', 'Requisitos que no tengo', 'Costo de la aplicación', 'Costo de la matricula', 'TOEFL', 'Idioma', 'Modalidad']
    resultado_final = pd.DataFrame(columns=columnas)

    # Iterar sobre las URLs y realizar web scraping
    for url in urls:
        dataframe_maestria = web_scraping(url)
        if dataframe_maestria is not None:
            # Agregar los datos al DataFrame final
            resultado_final = pd.concat([resultado_final, dataframe_maestria], ignore_index=True)

    # Pedir al usuario un nombre para el archivo Excel
    nombre_archivo = input('Ingresa el nombre del archivo Excel para guardar los datos: ')

    # Guardar el DataFrame combinado en un archivo Excel
    guardar_excel(resultado_final, nombre_archivo)
```
